## Question 1
#### Prepare a classification model using SVM for salary data 

```python
SD_train=pd.read_csv('/Users/apple/Downloads/SalaryData_Train(1).csv')
```

```python
SD_train
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
for column_name in SD_train.columns:
    if SD_train[column_name].dtype == object:
        SD_train[column_name] = le.fit_transform(SD_train[column_name])
    else:
        pass
```

```python
SD_train
```

```python
SD_train.info()
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x=norm_func(SD_train.iloc[:,0:13])
y=norm_func(SD_train.iloc[:,13])
```

```python
x
```

```python
x.shape
```

```python
y.shape
```

```python
clf = SVC(C= 15, gamma = 50)
clf.fit(x,y)
```

```python
SD_test=pd.read_csv('/Users/apple/Downloads/SalaryData_Test(1).csv')
```

```python
SD_test.info()
```

```python
for column_name in SD_test.columns:
    if SD_test[column_name].dtype == object:
        SD_test[column_name] = le.fit_transform(SD_test[column_name])
    else:
        pass
```

```python
X=norm_func(SD_test.iloc[:,0:13])
Y=norm_func(SD_test.iloc[:,13])
```

```python
X
```

```python
Y
```

```python
y_pred = clf.predict(X)
acc = accuracy_score(Y, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(Y, y_pred)
```

## Question 2
#### classify the Size_Category using SVM

```python
forest_fires=pd.read_csv('/Users/apple/Downloads/forestfires.csv')
```

```python
forest_fires
```

```python
forest_fires.info()
```

```python
for column_name in forest_fires.columns:
    if forest_fires[column_name].dtype == object:
        forest_fires[column_name] = le.fit_transform(forest_fires[column_name])
    else:
        pass
```

```python
X2=norm_func(forest_fires.iloc[:,0:30])
Y2=norm_func(forest_fires.iloc[:,30])
```

```python
X2
```

```python
Y2
```

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2,Y2, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.model_selection import KFold
```

```python
num_folds = 10
kfold = KFold(n_splits=10)
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=kfold)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,51,49,48,52],'C':[0.01,0.02,0.05,0.1,0.2] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 0.01, gamma = 50)
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
